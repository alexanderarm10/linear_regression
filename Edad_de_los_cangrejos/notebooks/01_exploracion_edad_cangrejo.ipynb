# 1. Carga de datos

```python
# Librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilos
sns.set(style='whitegrid')
```

```python
# Lectura de datos
df = pd.read_csv('../datos/crudos/CrabAgePrediction.csv')
df.info()
```

# 2. Distribución y relación

```python
# Distribución de la variable objetivo
df.groupby('Sex')['Age'].hist(bins=20);
```

```python
# Mínimo y máximo
print(f'Mínimo: {df.Age.min()}')
print(f'Máximo: {df.Age.max()}')
```

```python
# Corelaciones con pairplot
sns.pairplot(df, height=2);
```

```python
# Hetmap de la matríz de correlación
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
mc = df[columnas_numericas].corr()
sns.heatmap(mc, annot=True);
```

```python
# Se guardan los datos procesados
df.to_csv('../datos/procesados/edad_cangrejos.csv', index=False)
```
